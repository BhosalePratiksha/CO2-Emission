## Model Building

```python
x=df[["Volume","Weight"]].values
y=df["CO2"].values
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
weight=float(input("Enter the weight of the car in kg"))
volume=float(input("Enter the volume of the car in cm3"))
y_pred=lr.predict([[weight,volume]])
```

```python
print(f"The CO2 Emission of the car is:{y_pred}")
```
